```python
#Import pandas, read csv file and print the original result

import pandas as pd
df = pd.read_csv("C:\\Users\\rikua\\titanic_for_exercise.csv")
df

```

```python
# Clean the dataset
df = df.dropna()  # Drop rows with NaN values in columns
df = df[df['Fare'] > 0]  # Keep rows where fare is greater than zero
df = df[df['Sex'].isin(['male', 'female'])]  # Keep rows with 'male' or 'female' in 'Sex' column
df = df[df['Age'] >= 0]  # Keep rows where age is non-negative

# Check for non-finite values in 'Pclass' column
df = df.dropna(subset=['Pclass'])

#Age values and Siblings can be floats but for readability reasons I decided to change it into integers

df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)

# Convert age, Siblings/Spouses Aboard, and Parents/Children Aboard to integers

df['Age'] = df['Age'].astype(int)
df['Siblings/Spouses Aboard'] = df['Siblings/Spouses Aboard'].astype(int)
df['Parents/Children Aboard'] = df['Parents/Children Aboard'].astype(int)

# Output the cleaned dataset and the number of rows left
num_rows_left = len(df)
print("Number of rows left after changes:", num_rows_left)

print("\nOptimized Database:", df)
```
